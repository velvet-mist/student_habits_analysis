```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df= pd.read_csv("student_habits_performance.csv")
```

```python
df
```

```python
df.isnull().sum
```

```python
from sklearn.preprocessing import LabelEncoder

le= LabelEncoder()
columns_to_encode= ['gender', 'part_time_job','diet_quality','parental_education_level','internet_quality','extracurricular_participation']

for col in columns_to_encode:
    df[col]=le.fit_transform(df[col])

```

```python
df
```

```python
df=df.drop('student_id', axis=1)
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
```

```python
X= df.drop('exam_score', axis=1)
y= df['exam_score']
```

```python
from sklearn.feature_selection import SelectKBest, f_regression, mutual_info_regression

# Choose scoring function
selector = SelectKBest(score_func=f_regression, k=5)  # or mutual_info_regression

X_selected = selector.fit_transform(X, y)
selected_features = X.columns[selector.get_support()]
print("Selected features:", list(selected_features))
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

model = LinearRegression()
rfe = RFE(estimator=model, n_features_to_select=5)
rfe.fit(X, y)

selected_features = X.columns[rfe.support_]
print("Selected features:", list(selected_features))
```

```python
# by the two feature selections, the following columns will be used for analysis, 'study_hours_per_day', 
# 'social_media_hours', 'netflix_hours', 'sleep_hours', 'mental_health_rating'  and 'exercise frequency'
```

```python
X_selected= df[['study_hours_per_day', 'social_media_hours', 'netflix_hours', 'sleep_hours', 'mental_health_rating', 'exercise_frequency']]
```

```python
import matplotlib.pyplot as plt
X_selected.hist(bins=20, figsize=(12, 8))
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=X_selected)
plt.xticks(rotation=45)
plt.show()
```

```python
from sklearn.svm import SVR  # Support Vector Regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Scale features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
svm_model = SVR(kernel='rbf')  # You can also try 'linear', 'poly'
svm_model.fit(X_train_scaled, y_train)
```

```python
y_pred = svm_model.predict(X_test_scaled)

# Evaluation
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("R² Score:", r2)
```

```python
# Try different kernels
for kernel in ['linear', 'poly', 'rbf', 'sigmoid']:
    model = SVR(kernel=kernel)
    model.fit(X_train_scaled, y_train)
    y_pred = model.predict(X_test_scaled)
    print(f"{kernel} kernel → R²: {r2_score(y_test, y_pred):.3f}")
    print(f"{kernel} kernel -> MSE: {mean_squared_error(y_test, y_pred): .3f}\n")
```

```python
X_chosen = df[['study_hours_per_day']].values  # Use only one feature
y = df['exam_score'].values

scaler_X = StandardScaler()
scaler_y = StandardScaler()

X_scaled = scaler_X.fit_transform(X_chosen)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1)).ravel()

# Range for plotting
X_plot = np.linspace(X_scaled.min(), X_scaled.max(), 100).reshape(-1, 1)

kernels = ['linear', 'sigmoid']
plt.figure(figsize=(12, 4))

for i, kernel in enumerate(kernels):
    svr = SVR(kernel=kernel)
    svr.fit(X_scaled, y_scaled)
    y_pred = svr.predict(X_plot)

    plt.subplot(1, 2, i+1)
    plt.scatter(X_scaled, y_scaled, color='blue', label='Data')
    plt.plot(X_plot, y_pred, color='red', label=f'SVR ({kernel})')
    plt.title(f'{kernel.capitalize()} Kernel')
    plt.xlabel('Study Hours (scaled)')
    plt.ylabel('Exam Score (scaled)')
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
y_pred = svr_linear.predict(X_scaled)
print("R² Score:", r2_score(y_scaled, y_pred))
print("MSE:", mean_squared_error(y_scaled, y_pred))
```
